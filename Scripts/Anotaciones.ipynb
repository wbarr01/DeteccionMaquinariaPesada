<h1 style = 'color:blue'> Extraer data de las etiquetas y bbox presentes en las imágenes</h1>

```python
# Cargar librerias
import pandas as pd
import xml.etree.ElementTree as Xet
import os
from tqdm import tqdm_notebook

# Ruta donde se encuentra las anotaciones
anotc_root = '../Data/Coleccion-de-imagenes-master/Anotaciones'

# Extraer los labels y bbox de las imagenes y almacenarlas en un archivo csv
cols = ["filename", "width", "height", "depth", "label","pose","truncated","difficult","xmin","ymin","xmax","ymax"]
rows = []
  
# Parsing the XML file
#- Listar todos los archivos xml de la carpeta origen
fileList=os.listdir(anotc_root)
fileList.sort()
#- Estraer características del html
for f in tqdm_notebook(fileList):
    #- Nombre de la imágenes
    filename = f.replace('.xml', '.png')
    # Otras características
    xmlparse = Xet.parse(anotc_root+'\\'+f)
    root = xmlparse.getroot()
    size = root.find("size")
    width = size.find("width").text
    height = size.find("height").text
    depth = size.find("depth").text
    objects = root.findall("object")
    for i in objects:
        label = i.find("name").text
        truncated = i.find("truncated").text
        difficult = i.find("difficult").text
        box = i.find("bndbox")
        xmin = box.find("xmin").text
        ymin = box.find("ymin").text
        xmax = box.find("xmax").text
        ymax = box.find("ymax").text
        
        rows.append({"filename": filename,
                     "width": width,
                     "height": height,
                     "depth": depth,
                     "label": label,
                     "truncated": truncated,
                     "difficult": difficult,
                     "xmin": xmin,
                     "ymin": ymin,
                     "xmax": xmax,
                     "ymax": ymax
                     })
df_raw = pd.DataFrame(rows, columns=cols)
df_raw.to_csv('../Outputs/df_raw.csv', index = False)
df_raw.head()
```
